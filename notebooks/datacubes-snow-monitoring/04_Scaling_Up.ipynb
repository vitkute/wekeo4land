# Regional Analysis

In the previous notebooks, we explored how to query data for a single Area of Interest.

In this section, we aim to scale up the analysis and compare different regions. Since datacubes provide access to global datasets, scaling up is simply a matter of changing the area of interest. For our planned analysis, we will gather data for all of Austria, which covers an area of approximately 84,000 kmÂ². To put this into perspective, Austria is more than eight times larger than Northern Tyrol, meaning that for FSC and WDS, over 800 million unique data points need to be handled. Even analysis at this scale is possible with datacubes.

## Data Preparation


With Sentinel Hub, there are convenient APIs available to apply an analysis that works on a single geometry to multiple geometries. This makes it easy to scale up and fully utilize the potential of a datacube. These APIs are:

- Batch Processing API

- Batch Statistical API

For this use case, we will use the Batch Statistical API.

```python
import logging
import os

import pandas as pd
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt
from shapely.geometry.polygon import Polygon
from shapely.geometry.multipolygon import MultiPolygon

from sentinelhub import (
    DataCollection,
    SentinelHubBatchStatistical,
    SentinelHubStatistical,
    SHConfig,
    monitor_batch_statistical_job,
    Geometry,
    CRS
)
```

### Disclaimer

For this analysis, an Enterprise account on Sentinel Hub is necessary. You will also need access to an AWS S3 Bucket where the data will be saved.

More information about using Batch APIs can be found [here](https://docs.sentinel-hub.com/api/latest/api/batch-statistical/). Currently, the AWS integration in the sentinelhub package is not supported on the default EoXhub kernel. Therefore, the following cell cannot be run straight out of the box. However, when using the services locally outside of EDC, the cell can be run using the same credentials you would use on EDC ([see here](https://eurodatacube.com/documentation/credentials-on-edc)). Another option is to specify a custom kernel as described [here](https://eurodatacube.com/documentation/custom-jupyter-kernels).

Nonetheless, you can still run the analysis since the data generated through the requests is available in the notebook.

```python
from credentials import AWS_ID, AWS_SECRET

from sentinelhub.aws.batch import AwsBatchStatisticalResults
from sentinelhub.data_utils import get_failed_statistical_requests, statistical_to_dataframe

config = SHConfig()
config.sh_client_id = os.environ["SH_CLIENT_ID"]
config.sh_client_secret = os.environ["SH_CLIENT_SECRET"]
config.aws_access_key_id = AWS_ID
config.aws_secret_access_key = AWS_SECRET
```

### Loading area of interest

In this case we are loading a file containing the different states in Austria. 

```python
# Load in austrian states
states = gpd.read_file("../../data/raw/datacubes-snow-monitoring/input/austria_states.geojson")
# simplify geom to pass into payload 
states = states.set_geometry(states.simplify(100))
states["identifier"] = states.BL
crs_epsg = states.crs.to_epsg()
geom = states.geometry[0]
geom_sh = Geometry(geom, CRS(3035))
states.plot();
```

### Exporting Data for Use with the Statistical API

The Batch Statistical API operates asynchronously, running entirely on the cloud and saving the results there as well.

Therefore, the input data must also be accessible on the cloud. Here, we export the data in a format compatible with the Statistical API. This data is then uploaded to the cloud, where we will perform the analysis.

```python
states["geometry"] = [MultiPolygon([feature]) if isinstance(feature, Polygon) \
    else feature for feature in states["geometry"]]

gpkg = "../../data/raw/datacubes-snow-monitoring/input/austria_states.gpkg"
for geomtype in states.geom_type.unique():
   states[states.geom_type == geomtype].to_file(gpkg, driver="GPKG", layer=geomtype)
```

### Defining Data Sources

In this section, we define the data we want to use. For this example, we aim to generate maps of Persistent Snow Area and time series of Fractional Snow Cover for each state.

```python
config.sh_base_url = "https://creodias.sentinel-hub.com"
psa = DataCollection.define_byoc("da7e0012-8c43-42db-a5dc-cfd606c8b2dd")
gfsc = DataCollection.define_byoc("e0e66010-ab8a-46d5-94bd-ae5c750e7341")
config.sh_base_url = "https://services.sentinel-hub.com"
```

### Persistent Snow Area

For this analysis, we aim to obtain the following data:
- The persistent snow area in each region for each year

Using the Batch Statistical API offers several key advantages:

- Ability to specify a custom aggregation interval

- Automatic application to different polygons in the file

- Asynchronous computation entirely in the cloud

```python
with open("../../data/raw/datacubes-snow-monitoring/input/eval_psa_stat.js", "r") as file:
    evalscript = file.read()

aggregation = SentinelHubStatistical.aggregation(
    evalscript=evalscript,
    time_interval=("2016-01-01", "2022-01-01"),
    aggregation_interval="P1Y",
    resolution=(100, 100),
)

input_data=[
    SentinelHubStatistical.input_data(
        data_collection=psa,
    )
]
```

### Creating the request and logging the status

```python
INPUT_PATH = "s3://etc-datacubes-batch/austria_states.gpkg"
OUTPUT_PATH = "s3://etc-datacubes-batch/psa/"
input_features = SentinelHubBatchStatistical.s3_specification(INPUT_PATH, AWS_ID, AWS_SECRET)
output = SentinelHubBatchStatistical.s3_specification(OUTPUT_PATH, AWS_ID, AWS_SECRET)

client = SentinelHubBatchStatistical(config)
request = client.create(
    input_features=input_features,
    input_data=input_data,
    aggregation=aggregation,
    output=output,
    calculations={}
)
client.start_job(request)
```

```python
# To get the full output we must configure the logger first
logging.basicConfig(level=logging.INFO)

monitor_batch_statistical_job(request, config=config, analysis_sleep_time=60)
```

### Getting the data

```python
LOCAL_FOLDER = "output"

result_ids = range(1, 10) 
batch_results_request = AwsBatchStatisticalResults(
    request, data_folder=LOCAL_FOLDER, feature_ids=result_ids, config=config
)
results = batch_results_request.get_data(save_data=True, max_threads=4, show_progress=True)
df = statistical_to_dataframe(results)
df.to_csv("../../data/processing/datacubes-snow-monitoring/output/psa_states.csv")
```

### Analysis of Changes in Persistent Snow Area

First, let's examine the data returned from our request to the datacube.

```python
df = pd.read_csv("../../data/processing/datacubes-snow-monitoring/output/psa_states.csv", parse_dates=['interval_from'])
df.head()
```

Here, we prepare the data for analysis. From the statistical output, we calculate the number of pixels covered with persistent snow and derive the hectares covered.

To analyze the changes, we select two years to compare and compute the percentage change between them.

```python
year1 = 2018
year2 = 2021
df["year"] = df["interval_from"].dt.year
df["count_snow"] = df["persistent_snow_B0_mean"]*df["persistent_snow_B0_sampleCount"]
df["ha_snow"] = (df["count_snow"]).astype(int)
df["state"] = df["identifier"]
df_subset = df.iloc[np.isin(df["year"], [year1, year2])]
diff = pd.pivot_table(df_subset, values = "ha_snow", index = ["year", "state"], aggfunc=np.sum).unstack("year").reset_index()
diff.columns = ["state", year1, year2]
diff["percent_change"] = (diff[year2] / diff[year1])*100
diff
```

We can already see that the persistent snow area has increased in most regions. However, let's plot a map for a more visual representation:

```python
merged = states.merge(diff, left_on="BL", right_on="state")
fig, ax = plt.subplots(figsize=(15,4))
ax.set_title("Percentage change in persistent snow area from 2018-21");
merged.plot(column="percent_change", vmin=0, vmax=300, legend=True, ax=ax);
```

We can see that, except for Vorarlberg, all regions have increased their PSA during the analyzed time frame. A notable outlier is Styria, which shows an increase in PSA of over 3000%. However, since the initial total area was quite small, this relative change isn't very meaningful.

### Fractional Snow Cover Time Series

As another example, primarily to demonstrate the power of specifying a time interval for aggregation, we are analyzing differences in fractional snow cover across regions.

This analysis is simplified significantly by setting the aggregation interval to monthly, which is nearly all that's needed to obtain analysis-ready data. The steps for creating the request, monitoring progress, and retrieving the data remain the same.

In this case, we are obtaining monthly data for the 2017-2018 snow season.

```python
with open("../../data/raw/datacubes-snow-monitoring/input/eval_fsc_stat.js", "r") as file:
    evalscript_fsc = file.read()

aggregation_fsc = SentinelHubStatistical.aggregation(
    evalscript=evalscript_fsc,
    time_interval=("2017-11-01", "2018-06-01"),
    aggregation_interval="P1M",
    resolution=(100, 100),
)

input_data_fsc=[
    SentinelHubStatistical.input_data(
        data_collection=gfsc,
    )
]

OUTPUT_PATH_FSC = "s3://etc-datacubes-batch/fsc/"
output_fsc = SentinelHubBatchStatistical.s3_specification(OUTPUT_PATH_FSC, AWS_ID, AWS_SECRET)
```

### Creating the request and logging the status

```python
request_fsc = client.create(
    input_features=input_features,
    input_data=input_data_fsc,
    aggregation=aggregation_fsc,
    output=output_fsc,
    calculations={}
)
client.start_job(request_fsc)
```

```python
# To get the full output we must configure the logger first
logging.basicConfig(level=logging.INFO)

monitor_batch_statistical_job(request_fsc, config=config, analysis_sleep_time=60)
```

### Getting the data

```python
LOCAL_FOLDER = "output"

result_ids = range(1, 10) 
batch_results_request = AwsBatchStatisticalResults(
    request_fsc, data_folder=LOCAL_FOLDER, feature_ids=result_ids, config=config
)
results = batch_results_request.get_data(save_data=True, max_threads=4, show_progress=True)
```

### Analysis of Fractional Snow Cover Time Series

First, let's examine the data returned from our request to the datacube.

```python
df_fsc = statistical_to_dataframe(results)
df_fsc.to_csv("../../data/processing/datacubes-snow-monitoring/output/fsc_states.csv")
df_fsc.head()
```

Next, we prepare the data for analysis. From the statistical output, we calculate the number of pixels covered with snow each month and derive the number of square kilometers as well as the percentage of the geometry covered by snow.

```python
df_fsc = pd.read_csv("../../data/processing/datacubes-snow-monitoring/output/fsc_states.csv", parse_dates=['interval_from'])
df_fsc["month"] = df_fsc["interval_from"].dt.month
df_fsc["count_snow"] = df_fsc["fractional_snow_B0_mean"]*df_fsc["fractional_snow_B0_sampleCount"]
df_fsc["km2_snow"] = (0.1*0.1*df_fsc["count_snow"]).astype(int)
df_fsc["percent_covered"] = df_fsc["fractional_snow_B0_mean"]
df_fsc["state"] = df_fsc["identifier"]
```

Finally, we can plot the evolution of snow cover for different Austrian states during the 2017-2018 snow season.

```python
fsc = df_fsc[["percent_covered", "state", "month"]]
fsc_mth = pd.pivot_table(fsc, values="percent_covered", index=["state", "month"]).unstack(0)
fsc_mth.columns = fsc_mth.columns.droplevel()

months = np.arange(1,13)
idx = np.roll(months, 7)
replace = {A: B for A, B in zip(months, idx)}

fsc_mth["month"] = fsc_mth.index
fsc_mth.index = fsc_mth.index.astype(str)
fsc_mth["id"] = fsc_mth["month"].replace(replace)
fsc_mth.sort_values("id").plot(y=["Vorarlberg", "Steiermark", "Wien"], title="Percent of state covered by snow in the season 2017-18");
```

As expected, Vienna is seldom covered by snow. Additionally, in this season, Styria lost its snow cover earlier than Vorarlberg, with most snow being gone by March.

## Summary 

### Batch Statistical

- Provides a utility function to easily scale up analyses that already work on a small scale.

- Operates in the background, entirely on the cloud.

- Capable of handling any scale, including continental or global analyses.

### General Technical Summary

- No preprocessing necessary (resolution, time frame, AOI).

- Multiple data sources

- Application of high level concepts:
  - Spatial aggregation
  - Temporal aggregation

- Directly getting Analysis Ready Data

- Scaling up easily
